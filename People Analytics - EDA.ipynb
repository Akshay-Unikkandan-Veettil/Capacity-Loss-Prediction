```python
# See exact column names as PostgreSQL stored them

query_cols = """
    SELECT column_name 
    FROM information_schema.columns 
    WHERE table_name = 'employees'
    ORDER BY ordinal_position
"""
df_cols = pd.read_sql(query_cols, engine)
print(df_cols.to_string(index=False))
```

```python
#Comapany Attrition Percentage

query_attrition = """
    SELECT
        "Attrition",
        COUNT(*)                                AS employee_count,
        ROUND(COUNT(*) * 100.0 / SUM(COUNT(*))
        OVER(), 2)                              AS percentage
    FROM employees
    GROUP BY "Attrition"
    ORDER BY "Attrition" DESC
"""

df_attrition = pd.read_sql(query_attrition, engine)
print("Overall Attrition Rate:")
print(df_attrition.to_string(index=False))
```

```python
#Attrition by department

query_dept = """
    SELECT
        "Department",
        COUNT(*)                                        AS total_employees,
        SUM(CASE WHEN "Attrition" = 'Yes' THEN 1 
            ELSE 0 END)                                 AS leavers,
        ROUND(SUM(CASE WHEN "Attrition" = 'Yes' 
            THEN 1 ELSE 0 END) * 100.0 / 
            COUNT(*), 2)                                AS attrition_rate
    FROM employees
    GROUP BY "Department"
    ORDER BY attrition_rate DESC
"""

df_dept = pd.read_sql(query_dept, engine)
print("Attrition by Department:")
print(df_dept.to_string(index=False))

# Export
df_dept.to_csv(exports_path + "attrition_by_department.csv", index=False)
print("\n Exported!")
```

```python
#Attrition by Job Role

query_role = """
    SELECT
        "JobRole",
        COUNT(*)                                        AS total_employees,
        SUM(CASE WHEN "Attrition" = 'Yes' THEN 1 
            ELSE 0 END)                                 AS leavers,
        ROUND(SUM(CASE WHEN "Attrition" = 'Yes' 
            THEN 1 ELSE 0 END) * 100.0 / 
            COUNT(*), 2)                                AS attrition_rate
    FROM employees
    GROUP BY "JobRole"
    ORDER BY attrition_rate DESC
"""

df_role = pd.read_sql(query_role, engine)
print("Attrition by Job Role:")
print(df_role.to_string(index=False))

df_role.to_csv(exports_path + "attrition_by_jobrole.csv", index=False)
print("\n Exported!")
```

```python
#Attrition by Overtime

query_role = """
    SELECT
        "OverTime",
        COUNT(*)                                        AS total_employees,
        SUM(CASE WHEN "Attrition" = 'Yes' THEN 1 
            ELSE 0 END)                                 AS leavers,
        ROUND(SUM(CASE WHEN "Attrition" = 'Yes' 
            THEN 1 ELSE 0 END) * 100.0 / 
            COUNT(*), 2)                                AS attrition_rate
    FROM employees
    GROUP BY "OverTime"
    ORDER BY attrition_rate DESC
"""

df_role = pd.read_sql(query_role, engine)
print("Attrition by Over Time:")
print(df_role.to_string(index=False))

df_role.to_csv(exports_path + "attrition_by_jobrole.csv", index=False)
print("\n Exported!")
```

```python
#Attrition by Job Level

query_role = """
    SELECT
        "JobLevel",
        COUNT(*)                                        AS total_employees,
        SUM(CASE WHEN "Attrition" = 'Yes' THEN 1 
            ELSE 0 END)                                 AS leavers,
        ROUND(SUM(CASE WHEN "Attrition" = 'Yes' 
            THEN 1 ELSE 0 END) * 100.0 / 
            COUNT(*), 2)                                AS attrition_rate
    FROM employees
    GROUP BY "JobLevel"
    ORDER BY attrition_rate DESC
"""

df_role = pd.read_sql(query_role, engine)
print("Attrition by Job Level:")
print(df_role.to_string(index=False))

df_role.to_csv(exports_path + "attrition_by_jobrole.csv", index=False)
print("\n Exported!")
```

```python
#Attrition by Salaries 

query_salary = """
    SELECT
        "Attrition",
        ROUND(AVG("MonthlyIncome")::numeric, 2)     AS avg_salary,
        ROUND(MIN("MonthlyIncome")::numeric, 2)     AS min_salary,
        ROUND(MAX("MonthlyIncome")::numeric, 2)     AS max_salary,
        ROUND(PERCENTILE_CONT(0.5) WITHIN GROUP 
            (ORDER BY "MonthlyIncome")::numeric, 2) AS median_salary
    FROM employees
    GROUP BY "Attrition"
    ORDER BY "Attrition" DESC
"""

df_salary = pd.read_sql(query_salary, engine)
print("Salary Analysis by Attrition:")
print(df_salary.to_string(index=False))

df_salary.to_csv(exports_path + "salary_by_attrition.csv", index=False)
print("\n Exported!")
```

```python
#Attrition by Tenure

query_tenure = """
    SELECT
        "Attrition",
        ROUND(AVG("YearsAtCompany")::numeric, 2)        AS avg_years_company,
        ROUND(AVG("YearsSinceLastPromotion")::numeric, 2) AS avg_years_no_promotion,
        ROUND(AVG("YearsWithCurrManager")::numeric, 2)  AS avg_years_with_manager,
        ROUND(AVG("TotalWorkingYears")::numeric, 2)     AS avg_total_experience
    FROM employees
    GROUP BY "Attrition"
    ORDER BY "Attrition" DESC
"""

df_tenure = pd.read_sql(query_tenure, engine)
print("Tenure Analysis by Attrition:")
print(df_tenure.to_string(index=False))

df_tenure.to_csv(exports_path + "tenure_by_attrition.csv", index=False)
print("\n Exported!")
```

```python
query_flightrisk = """
    SELECT
        "Department",
        "JobRole",
        "JobLevel",
        "MonthlyIncome",
        "OverTime",
        "YearsSinceLastPromotion",
        "YearsAtCompany",
        "YearsWithCurrManager",
        "JobSatisfaction",
        "WorkLifeBalance",
        "EnvironmentSatisfaction",
        "Attrition",
        CASE
            WHEN "OverTime" = 'Yes' 
             AND "JobLevel" = 1          THEN 'High'
            WHEN "OverTime" = 'Yes' 
             AND "JobSatisfaction" <= 2  THEN 'High'
            WHEN "JobLevel" = 1 
             AND "JobSatisfaction" <= 2  THEN 'High'
            WHEN "OverTime" = 'Yes'      THEN 'Medium'
            WHEN "JobLevel" <= 2         THEN 'Medium'
            WHEN "JobSatisfaction" <= 2  THEN 'Medium'
            ELSE                              'Low'
        END AS risk_tier
    FROM employees
    ORDER BY risk_tier ASC
"""

df_flightrisk = pd.read_sql(query_flightrisk, engine)
print("Flight Risk Distribution:")
print(df_flightrisk['risk_tier'].value_counts())

df_flightrisk.to_csv(exports_path + "flight_risk_employees.csv", index=False)
print("\nExported!")
```
